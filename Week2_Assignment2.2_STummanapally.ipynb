```python
#from __future__ import print_function, division
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# Loading Covid 19 cases data to a dataframe.
Covid_df = pd.read_csv("Data/CovidCases.csv", sep=',')

Covid_df.head()     #Getting few rows from the dataframe.
```

This is the dataset which contains the info of number of Covid-19 cases that were confirmed from the day 1 of the pandemic affect in India. The data contains the state-wise data on a daily basis. We will learn more about the key variable we will be using for our research question in the graphs below.

## Identify a question or question(s) that you would like to explore in your data set.

Research Question/Topic:
To understand the impact of Covid-19 pandemic in India, I want to research the trend in increase in the number of infected people across the initial months and the most impacted state in India.


## Create at least three graphs that help answer these questions. Make sure your graphs are clearly readable and are labeled appropriately and professionally.

```python
Covid_df['Date'] =  pd.to_datetime(Covid_df['Date'])
Covid_df.groupby('Date')['New cases'].sum().plot(kind='line')
plt.gca().set_xbound(dt.date(2020, 1, 30), dt.date(2020, 8, 30))
plt.title('Daily New Cases in India')          #Adding a title to the graph
plt.xlabel('Date')                             #Labeling the axis to the graph
plt.ylabel('Number of new cases')
plt.show()
```

```python
Covid_df.groupby('Name of State / UT')['Total Confirmed cases'].max().plot(kind='bar')
plt.title('Total Confirmed Cases - Statewise')          #Adding a title to the graph
plt.xlabel('State')                                     #Labeling the axis to the graph
plt.ylabel('Total Confirmed Cases')
plt.show()
```

```python
Covid_df['Death'] =  pd.to_numeric(Covid_df['Death'])
Covid_df.groupby('Name of State / UT')['Death'].max().plot(kind='bar')
plt.title('Total Deaths - Statewise')          #Adding a title to the graph
plt.xlabel('State')                            #Labeling the axis to the graph
plt.ylabel('Total Deaths')
plt.show()
```

## Explain what you have learned from each of your graphs.

Graph 1 : Daily New Cases in India - This shows us the daily trend of increase in the infected people in india. From Febraury to August, the number of new covid-19 cases increase exponentially in July.

Graph 2 : Total Confirmed Cases - Statewise - This shows us the areas or states that are mostly impacted in this pandemic. It looks like during the peroid of February to August, the state of Maharastra has been affected by Covid-19 more than any other state.

Graph 3 - Total Deaths - Statewise - This shows us the areas or states that had most deaths in this pandemic. It looks like during the peroid of February to August, the state of Maharastra has deaths due to Covid-19 more than any other state.
